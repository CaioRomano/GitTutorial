# Git e Github

## Verificar versão do Git no computador

Para verificar a versão do git, basta digitar no seu terminal do computador, ou no terminal localizado em sua IDE, o comando `git --version`

## Repositórios

Para criar um repositório, basta digitar o comando, no terminal, na raiz de seu projeto: `git init`

### Mudança do nome padrão do branch principal

A fim de alterar o nome do branch principal do seu repositório, para ficar de acordo com o padrão estabelecido, você deverá digitar o comando: `git config --global init.defaultBranch [nome novo a ser inserido que, neste caso, será 'main']`

Outra forma de alterar o nome de uma branch é o comando: `git branch -M [novo nome a ser inserido]`

Uma branch padrão é criada automaticamente pelo git, todas as vezes que você digita o comando `git init`.

### Branches

Branches são espaços paralelos do seu projeto. Cada branch pode armazenar uma determinada versão do seu código/projeto sem intereferir na outra versão. Isso se deve ao paralelismo dos branches.

### Comando `git status`

Este comando mostra ao usuário se existem commits pendentes, quais arquivos são novos, se houve modificação ou exclusão de algum arquivo. Além disso, ele mostrará arquivos não rastreados, ou seja, arquivos que não estão sendo monitorados pelo git para controle de versão. Basta digitar, em seu terminal, o comando `git status`.

### Comando `git add`

Este comando é responsável por adcionar arquivos, pastas, etc para controle de versão do git. A partir disso, o arquivo em questão será alocado para commit. Esta ação é basicamente dizer ao git que você quer adicionar o arquivo x para controle de versão. O comando para fazer isso é: `git add [nome do arquivo]`

### Comando `git commit -m "[mensagem acerca do commit]"`

O comando acima realiza o commit do arquivo em questão para que seja , de fato, rastreado pelo git, cirando uma espécie de cópia do mesmo, de modo que, quando for necessário, o usuário poderá voltar para a versão em questão, onde o arquivo possuía um determinado formato antes de falhar.

### OBSERVAÇÃO OBRIGATÓRIA

É um padrão realizar commits regularmente em seu projeto. Isso faz com que, caso o sistema falhe, ou quebre, o usuário sempre poderá voltar para uma versão anterior do sistema antes da falha.

### Comando `git add ./`

O comando acima automatiza a adição dos arquivos de seu projeto não rastreados pelo git de uma só vez, ao invés de adicionar um arquivo por vez de maneira manual. Basta digitar o `git add ./`

### Comando `git log`

Esse comando retorna um histórico de todos os commits feitos pelo usuário, bem como quem fez o commit e qual a mensagem de cada commit.

### Comando `git checkout [hash do commit no qual você quer voltar]`

O comando acima pode fazer com que o usuário retorne a um commit anterior ao atual. Caso haja um evento onde seu arquivo possua bugs, é mais aconselhável voltar para uma versão onde não havia o erro. Para isso, além de digitar o comando acima, na parte do hash do commit, você consegue o hash do commit desejado através do comando `git log`.

Para voltar ao estado mais atual do nosso código, ou ao commit mais atual, basta digitar `git checkout main`

### Vinculando repositório local ao repositório remoto

Ao digitar o comando `git remote add origin [link do repositório remoto]` você estará adicionando seu repositório local ao repositório remoto, que está localizado nos servidores do github.

#### Comando `git push -u origin main`

O comando acima leva ao repositório remoto todas as alterações feitas no repositório local, de modo que o origin é um alias do repositório remoto, isso é feito para evitar de sempre enviar o link do repositório remoto.

### Branches

Uma branch no git representa uma versão do proejto atual, na qual a partir de um determinado ponto é criado uma cópia do projeto, mas todas as alterações feitas nessa ramificação não afeta na branch principal. Isso é útil para implementar funcionalidades novas ao código e o desenvolvedor quer evitar de quebrar o projeto completamente. Portanto, caso o sistema quebre, não será no projeto principal, mas numa cópia do mesmo, que existe de modo paralelo ao projeto principal.

#### Comando `git branch [nome da branch]`

Esse comando permite criar novas branches ao projeto, mas altera a branch em que o usuário está.

Caso queira ir para uma branch específica, basta digitar o comando `git checkout [nome da branch]`

Agora, se você quer criar uma nova branch e entrar nela imediatamente após sua criação, o comando `git checkout -b [nome da branch]`

#### Importante!!!

As branches no git são criadas a partir do ponto em que se encontra o projeto. E caso você quiser criar uma nova branch filha de uma branch filha, essa nova branch herdará todo o código contido na branch filha e não na branch pai, ou branch principal.

#### Comando para criar múltiplas branches de uma vez

A fim de evitar a criação manual de diversas branches, você pode evitar isso ao digitar o seguinte comando: `git branch [nome da branch] [nome da branch herdada]`

#### Comando `git push --set-upstream origin [nome da branch]`

Esse comando faz o seguinte: Ele diz ao repositório remoto que ele receberá as alterações de determinado branch especificado pelo usuário. A partir daí, todas as alterações feitas naquele branch serão refletidas no branch remoto.

Outra forma de realizar a mesma operação é digitando o comando `git push -u origin [nome da branch]`

Depois de digitar esse comando, nas próximas alterações que você fizer naquele branch, basta digitar o comando `git push` para refletir tais mudanças no repositório remoto.

#### Comando `git pull`

Caso hajam alterações no repositório remoto, referentes a uma determinada branch, e você quer atualizar o repositório local com tais alterações.

### Merging

Quando um usuário termina de adicionar determinadas funcionalidades numa branch específica e quer adicionar as mesmas na branch principal, ele deve, primeiro, realizar um "pull request". Isso significa copartilhar suas modificações com todo o seu time, para revisão e feedback.

Caso sua modificação seja aprovada e mesclada, ela se tornará parte do projeto principal.

Essa prática permite um código estável e organizado.

### Merge Conflicts

Merge conflicts são conflitos onde o git não sabe exatamente o que fazer quando há alterações feitas em diferentes branches no mesmo arquivo. Por isso, o git deixa a cargo do usuário sobre qual modificação deve ser persistida e qual deve ser apagada, permitindo, também, a opção de manter determinados trechos de ambas as alterações no mesmo arquivo.

### Comando `git reset`

Existe o comando `git checkout [hash do commit]` no qual permite ao usuário voltar a um commit específico. Mas, e se nós quiséssemos voltar a um commit específico e excluir tudo o que veio depois? O comando acima resolve esse problema.

Este comando permite que o usuário retorne a um commit específico, excluindo todos os commits feitos após o mesmo. Ademais, retorna todas as alterações feitas nesses commits, deixando a cargo do usuário sobre o que fazer com eles: Manter as alterações feitas, e escolher quais delas manter, ou simplesmente excluir tudo.

Existe 3 tipos de `git reset`: O soft reset, o mixed reset e o hard reset

#### Soft Reset

Neste tipo de comando, você simplesmente move o commit atual para o lugar mais atual, ou seja, ele agora é como se fosse o commit mais recente. Enquanto que o conteúdo dos outros resets posteriores, são colocados em modo "staging". Isso significa que as alterações feitas nesses commits estão salvas num diretório, como se elas estivessem sendo rastreadas pelo git, como quando acontece quando queremos commitar arquivos. Em resumo, isso exclui commits posteriores, sem excluir as mudanças que acontecerem nesses commits

Para realizar o soft reset, é preciso fazer o seguinte: `git reset --soft [hash do commit]`

#### Mixed Reset

No mixed reset, basta digitar o comando padrão: `git reset [hash do commit]`. Neste tipo de reset, acontece de maneira similar ao soft reset, porém, as alterações feitas, não ficam no modo staging, mas o usuário deverá adicionar os arquivos para staging manualmente, através do comando `git add`. Apesar de tudo, todas as alterações estão alocadas ainda no diretório de trabalho, não mais no modo staging

#### Hard Reset

Neste modo, ele remove todas os commits posteriores, e todo o seu conteúdo, nem mesmo mantendo as alterações no modo staging ou no diretório de trabalho. É uma limpeza completa de conteúdo, como se os commits não existissem.

Para fazer isso, basta digitar: `git reset --hard [hash do commit]`

### Comando `git revert`

Esse comando é usado quando há um problema no código, durante a produção, e você quer retornar a um commit específico, sem remover os outros commits que vieram depois, é útil quando se quer manter um registro claro de todas as alterações feitas.

Ao executar `git revert [hash do commit]`, será feito uma espécie de merge dos commits mais recentes até o commit em que você quer. Depois disso, cabe ao usuário escolher o que excluir e o que manter.

Então, deverá inserir o(s) arquivo(s) alterado(s) para a área de staging, com o comando `git add ./`. E, por fim, digitar o comando `git revert --continue` para que o processo de reversão seja finalizado.

A seguir, existem duas opções que podem ser feitas. A primeira é inserir uma mensagem para o processo de reversão (semelhante a mensagem do comando `git commit`, ou apenas digitar no terminal o comando: `:qa!`

### Comando `git stash`

Esse comando serve para realizar uma tarefa diferente da que você está fazendo atualmente, e quer guardar seu trabalho num outro lugar, seguro, e voltar para ele depois de terminar a nova tarefa.

Ele salva todas as alterações feitas, sejam elas no modo staging ou fora desse modo.

Para salvar seu trabalho antigo e focar no atual, deverá digitar `git stash`

Depois que você terminar a nova tarefa, você vai querer retornar àquela antiga, certo? Bom, para fazer isso, você necessita digitar o comando: `git stash list`. Com ele, você encontrará o stash que está armazenando seu trabalho anterior.

Em seguida, você deve copiar o código de stash e digitar o comando `git stash apply [nome do stash]`

O resultado será que tanto o trabalho atual quanto o anterior estão no mesmo lugar, resultando numa espécie de merge conflict. Mas, para resolver, basta excluir e manter as alterações que desejar, de modo manual.

## Observação Importante!!!

Para nomear commits a serem enviadas para o repositório remoto, ou local, você deve enviar uma mensagem para cada commit feito, correto? Bom, o padrão de mensagem a ser inserida é sempre adicionar um verbo no infitivo seguido do objeto de ação desse verbo.

Por exemplo:

`git commit -m "adicionar Readme.md"`

`git commit -m "alterar código nos arquivos"`
